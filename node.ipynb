```python
import openpyxl
import matplotlib.pyplot as plt

# Excelファイルを読み込む
modeset =['AIDS','COLLAB','IMDB','PCM','PDBS','PPI','REDDIT']
mode = modeset[0]
filename = 'CTsearch\\'+ mode+'.xlsx'
book = openpyxl.load_workbook(filename)
sheet_names = book.sheetnames
print(sheet_names)

# Excelシートからx軸とy軸のデータを読み込む
ws = book[sheet_names[0]]        
x_data = [cell.value for cell in ws['A'][1::]]
y_data = [cell.value for cell in ws['D'][1::]]

print(x_data)
print(y_data)
# 折れ線グラフを描画する
plt.plot(x_data, y_data)

# グラフのタイトル、軸ラベルを設定する
plt.title('')
plt.xlabel('Number of Graphs',fontsize=20)
plt.ylabel('Filtering Time',fontsize=20)

plt.ylim(0,max(y_data))
plt.xlim(0, max(x_data))
# plt.yscale('log')
savefile = 'compare_filtertime\\'+mode+'.eps'
plt.savefig(savefile)

plt.show()
```
