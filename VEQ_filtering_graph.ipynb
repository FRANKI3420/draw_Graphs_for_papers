```python
import openpyxl
import matplotlib.pyplot as plt
import numpy as np

# Excelファイルを読み込む
plt.figure(figsize=(8,5))  # 横幅: 10インチ, 縦幅: 6インチ

modeset =['AIDS','COLLAB','IMDB-MULTI','pcms','pdbs','ppigo','REDDIT-MULTI-5K']
mode = modeset[0]
type = '_R'
filename = 'VEQ_filtering_graph10\\'+ mode+'.xlsx'
book = openpyxl.load_workbook(filename)
sheet_names = book.sheetnames
print(sheet_names)

# Excelシートからx軸とy軸のデータを読み込む
ws = book[sheet_names[0]]        
x_data = [cell.value for cell in ws['A'][1::]]
r4_data = [cell.value for cell in ws['B'][1::]]
r8_data = [cell.value for cell in ws['C'][1::]]
r16_data = [cell.value for cell in ws['D'][1::]]
r32_data = [cell.value for cell in ws['E'][1::]]
r64_data = [cell.value for cell in ws['F'][1::]]

x_data.insert(0,0)
r4_data.insert(0,0)
r8_data.insert(0,0)
r16_data.insert(0,0)
r32_data.insert(0,0)
r64_data.insert(0,0)

print(x_data)
print(r4_data)
print(max(r4_data))
print(max(r64_data))


# 折れ線グラフを描画する
plt.plot(x_data, r4_data, label='4R')
plt.plot(x_data, r8_data, label='8R')
plt.plot(x_data, r16_data, label='16R')
plt.plot(x_data, r32_data, label='32R')
plt.plot(x_data, r64_data, label='64R')


# グラフのタイトル、軸ラベルを設定する
plt.title('')
plt.xlabel('|G|',fontsize=20)
plt.ylabel('Filtering Time',fontsize=20)

plt.ylim(0,max(r4_data)+0.5)
plt.xlim(0, max(x_data))
xticks = np.linspace(0, 40000, 5)  # X軸の目盛り位置を計算
plt.xticks(xticks,fontsize=15)
plt.yticks(fontsize=15)


plt.legend(loc='upper center',bbox_to_anchor=(0.5, 1.11),ncol =10,fontsize=12) #凡例

# plt.yscale('log')
savefile = 'VEQ_fil_graph\\'+mode+type+'.eps'
plt.savefig(savefile)

plt.show()
```
