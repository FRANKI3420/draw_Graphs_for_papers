```python
#Filtering Precision AIDS
import matplotlib.pyplot as plt
import numpy as np
import openpyxl

book = openpyxl.load_workbook('AIDSnode.xlsx',data_only=True)
index = 0

fig, ax1 = plt.subplots()


for sheet in book.worksheets:
    x_data = [cell.value for cell in sheet['A'][1::]]
    y_data = [cell.value for cell in sheet['C'][1::]]
    z_data = [cell.value for cell in sheet['B'][1::]]

print(x_data)
print(y_data)
print(z_data)


# グラフのタイトル、軸ラベルを設定する
ax1.set_title('')
ax1.set_xlabel('|G|',fontsize=20)
ax1.set_ylabel('Inclusion Check Time',fontsize=20)

ax1.set_ylim(0, 70)
ax1.set_xlim(0, 40000)

# 折れ線グラフを描画する
ax1.plot(x_data, y_data)

ax2 = ax1.twinx()
# ax1.plot(x_data, z_data)
ax2.bar(x_data, z_data)

 
```
