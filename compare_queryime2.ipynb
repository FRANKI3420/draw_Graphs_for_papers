```python
import openpyxl
import matplotlib.pyplot as plt
import numpy as np

# Excel
plt.figure(figsize=(8,5.2))  # : 10, : 6

modeset =['AIDS','COLLAB','IMDB','PCM','PDBS','PPI','REDDIT']
mode = modeset[0]
type = '_R'
filename = 'compare_queryime.xlsx'
book = openpyxl.load_workbook(filename,data_only=True)

# Excelxy
ws = book[mode]        
x_data = [cell.value for cell in ws['A'][1::]]
r4_data = [cell.value for cell in ws['O'][1::]]
r8_data = [cell.value for cell in ws['P'][1::]]
r16_data = [cell.value for cell in ws['Q'][1::]]
r32_data = [cell.value for cell in ws['R'][1::]]
r64_data = [cell.value for cell in ws['S'][1::]]

print(x_data)

# 
plt.plot(x_data, r4_data, label='4R')
plt.plot(x_data, r8_data, label='8R')
plt.plot(x_data, r16_data, label='16R')
plt.plot(x_data, r32_data, label='32R')
plt.plot(x_data, r64_data, label='64R')


# 
plt.title('')
plt.xlabel('|G|',fontsize=20)
plt.ylabel('Query Time',fontsize=20)

plt.ylim(0,1)
plt.xlim(0, 40000)
xticks = np.linspace(0, 40000, 5)  # X
plt.xticks(xticks,fontsize=15)
plt.yticks(fontsize=15)
# plt.yscale('log')


plt.legend(loc='upper center',bbox_to_anchor=(0.5, 1.11),ncol =10,fontsize=12) #

# plt.yscale('log')
savefile = 'compare_querytime2\\'+mode+type+'.eps'
plt.savefig(savefile)

plt.show()
```
