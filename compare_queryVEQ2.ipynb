```python
import openpyxl
import matplotlib.pyplot as plt
import numpy as np

# Excelファイルを読み込む
plt.figure(figsize=(8,5.2))  # 横幅: 10インチ, 縦幅: 6インチ

modeset =['AIDS','COLLAB','IMDB','PCM','PDBS','PPI','REDDIT']
mode = modeset[0]
type = '_R'
filename = 'VeQ2.xlsx'
book = openpyxl.load_workbook(filename,data_only = True)

# Excelシートからx軸とy軸のデータを読み込む
ws = book[mode]        
x_data = [cell.value for cell in ws['A'][1::]]
r4_data = [cell.value for cell in ws['G'][1::]]
r8_data = [cell.value for cell in ws['H'][1::]]
r16_data = [cell.value for cell in ws['I'][1::]]
r32_data = [cell.value for cell in ws['J'][1::]]
r64_data = [cell.value for cell in ws['K'][1::]]

print(x_data)
print(r4_data)

# 折れ線グラフを描画する
plt.plot(x_data, r4_data, label='4R')
plt.plot(x_data, r8_data, label='8R')
plt.plot(x_data, r16_data, label='16R')
plt.plot(x_data, r32_data, label='32R')
plt.plot(x_data, r64_data, label='64R')


# グラフのタイトル、軸ラベルを設定する
plt.title('')
plt.xlabel('|G|',fontsize=20)
plt.ylabel('Query Time',fontsize=20)

x  =max([x for x in x_data if x is not None])

plt.ylim(0,1)
plt.xlim(0, x)
xticks = np.linspace(0, x, 3)  # X軸の目盛り位置を計算
plt.xticks(xticks,fontsize=15)
yticks = np.linspace(0, 1, 5)  # X軸の目盛り位置を計算
plt.yticks(yticks,fontsize=15)
# plt.yscale('log')


plt.legend(loc='upper center',bbox_to_anchor=(0.5, 1.11),ncol =10,fontsize=12) #凡例

# plt.yscale('log')
savefile = 'compare_queryVEQ2\\'+mode+type+'.eps'
plt.savefig(savefile)

plt.show()
```
