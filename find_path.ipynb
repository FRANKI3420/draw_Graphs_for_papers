## 経路列挙とパスの長さ(重みの合計)

```python
def path_total_weight(path, edge_labels):
    total = 0
    for i in range(len(path)-1):
        total += edge_labels[(path[i],path[i+1])]
    return total
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# ファイルを読み込む
with open("query.txt", "r") as f:
    data = f.readlines()

# 頂点数を取得
n = int(data[0])

# 頂点IDを取得
nodes = [int(x) for x in data[1:n+1]]

# グラフを初期化
G = nx.Graph()
for i in range(n):
    node = nodes[i]
    # 頂点IDをラベルとして設定
    G.add_node(node, label=node)
    # 頂点の位置を指定
    if not 'pos' in G.nodes[node]:
        G.nodes[node]['pos'] = (i, 0)
        
# 頂点をグラフに追加
for i in range(n):
    node = nodes[i]
    # 頂点IDをラベルとして設定
    G.add_node(node, label=node)
    # 頂点の位置を指定
    G.nodes[node]['pos'] = (i, 0)
    

# 辺の数を取得
m = int(data[n+1])

# 辺をグラフに追加
for line in data[n+2:]:
    u, v = map(int, line.split())
    G.add_edge(u, v)

# 頂点の位置を修正するためのforce-directed layoutアルゴリズムを使用
pos = nx.spring_layout(G, pos={node:(node_data['pos'][0], 0) for node, node_data in G.nodes(data=True)}, fixed=G.nodes())

# グラフを描画
nx.draw(G, pos, labels={node:node_data['label'] for node, node_data in G.nodes(data=True)}, with_labels=True)

# 描画したグラフを表示
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# ファイルからグラフを構築
def read_graph(file_path):
    with open(file_path, "r") as f:
        n = int(f.readline().rstrip())
        nodes = f.readline().rstrip().split()
        G = nx.Graph()
        G.add_nodes_from(nodes)

        m = int(f.readline().rstrip())
        for i in range(m):
            u, v = f.readline().rstrip().split()
            G.add_edge(u, v)

    return G

# 頂点の座標を設定
def set_node_positions(G):
    pos = nx.spring_layout(G)
    for node in G.nodes():
        G.nodes[node]['pos'] = pos[node]

# グラフを描画
def draw_graph(G):
    nx.draw_networkx(G, pos=nx.get_node_attributes(G, 'pos'))
    plt.show()

if __name__ == "__main__":
    file_path = "query.txt"
    G = read_graph(file_path)
    set_node_positions(G)
    draw_graph(G)
```
