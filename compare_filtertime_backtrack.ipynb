```python
#Filtering Precision AIDS
import matplotlib.pyplot as plt
import numpy as np
import openpyxl

plt.figure(figsize=(8,5))  # 横幅: 10インチ, 縦幅: 6インチ



# 棒の配置位置、ラベルを用意
hatch_list = ['/', 'O']

plt.figure(figsize=(8,5))  # 横幅: 10インチ, 縦幅: 6インチ


modeset =['AIDS','COLLAB','IMDB','PCM','PDBS','PPI','REDDIT']
mode = modeset[0]

# for mode in modeset:
  
print(mode)
  
# 各系列のデータを用意
CT = []
CT_back  =[]

CT_result = []
CT_back_result  =[]


x = np.array([1, 2])
labels = ["4R", "4B"]
  
#CT
book = openpyxl.load_workbook('CT3.xlsx',data_only=True)
ws = book[mode]
index =0
column = ws['F']  # C列を取得

for cell in column:
    index += 1
    if cell.value is not None :
        if index!=1:# セルの値が空でない場合に値を配列に追加
            CT.append(cell.value)
    else:  # セルの値が空になったらループを抜ける
        break
        
CT_result.append(CT[0])
CT_result.append(CT[5])
print(CT_result)

#CT
filename = "nonback_search\\" + mode +"_result.xlsx"
book = openpyxl.load_workbook(filename,data_only=True)
sheetname = mode+"_result"
ws = book[sheetname]
index =0
column = ws['F'] 

for cell in column:
    index += 1
    if cell.value is not None :
        if index!=1:# セルの値が空でない場合に値を配列に追加
            CT_back.append(cell.value)
    else:  # セルの値が空になったらループを抜ける
        break

CT_back_result.append(CT_back[0])
CT_back_result.append(CT_back[5])
print(CT_back_result)

data = [CT_back_result,CT_result]

# マージンを設定
margin = 0.4  #0 <margin< 1
totoal_width = 1 - margin

index = 0
# 棒グラフをプロット
for i, h in enumerate(data):
  pos = x - totoal_width *( 1- (2*i+1)/len(data) )/2
  if index%len(data)==0:
    color = "lightgray"
    name = "backtrack no"
  elif index%len(data)==1:
     color = "WHITE"
     name = "backtrack ok"

  plt.bar(pos, h, width = totoal_width/len(data),edgecolor="black",facecolor=color, hatch = hatch_list[index%len(data)],label=name)
  index+=1
  
# ラベルの設定
plt.legend(loc='upper center',bbox_to_anchor=(0.5, 1.14),ncol =len(data),fontsize=15) #凡例
plt.xticks(x, labels, fontsize=20)
plt.ylim(0.01, 1)
plt.yticks(fontsize=20)  # y軸のメモリの文字サイズを設定
plt.yscale('log')
plt.ylabel("Filtering Time (ms)", fontsize=20)
# plt.subplots(figsize=(3.94, 1.97), dpi=100)
filename ='compare_filtertime_backtrack//'+mode+'.eps'
plt.savefig(filename)
```
